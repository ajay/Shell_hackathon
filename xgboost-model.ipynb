```python
import pandas as pd
import xgboost as xgb
from sklearn.metrics import mean_squared_error
import os
```

```python

data = pd.read_csv("/kaggle/input/biomass-dist/Biomass_History.csv")
features = data[["Index","Latitude","Longitude","2010","2011","2012","2013","2014","2015","2016"]]
target = data[["2017"]]

model = xgb.XGBRegressor()

model.fit(features,target)

result = model.predict(features)

print(result)
```
