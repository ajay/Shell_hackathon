```python
import pandas as pd
import xgboost as xgb
from sklearn.metrics import mean_squared_error
import os
```

```python

data = pd.read_csv("/kaggle/input/new-test/Untitled spreadsheet - Sheet1.csv")
features = data[["Index","Latitude","Longitude","2010","2011","2012","2013","2014","2015","2016"]]
target = data[["2017"]]

model = xgb.XGBRegressor()

model.fit(features,target)

result = model.predict(features)
final_2017 = pd.DataFrame({"Predicted_2017_Biomass": result})
final_2017.to_csv("/kaggle/working/predicted_biomass_2017.csv", index=False)

print(result)
```

```python
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split

data = pd.read_csv("/kaggle/input/2018-inc-dataset/2018_inc_dataset - Sheet1.csv")

features = data[["Index", "Latitude", "Longitude", "2010", "2011", "2012", "2013", "2014", "2015", "2016","2017"]]
target = data["2018"]

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)


model = xgb.XGBRegressor()
model.fit(X_train, y_train)


features_2019 = data[["Index", "Latitude", "Longitude", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018"]]


predictions_2019 = model.predict(features_2019)


final_2019 = pd.DataFrame({"Predicted_2019_Biomass": predictions_2019})

final_2019.to_csv("/kaggle/working/predicted_biomass_2019.csv", index=False)

print(final_2019)
```

```python
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split


data = pd.read_csv("/kaggle/input/2018-inc-dataset/2018_inc_dataset - Sheet1.csv")


features = data[["Index", "Latitude", "Longitude", "2010", "2011", "2012", "2013", "2014"]]
target = data["2015"]


X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)


model = xgb.XGBRegressor()
model.fit(X_train, y_train)


features_2016 = data[["Index", "Latitude", "Longitude", "2011", "2012", "2013", "2014", "2015"]]


predictions_2016 = model.predict(features_2016)


final_2016 = pd.DataFrame({"Predicted_2016_Biomass": predictions_2016})

final_2016.to_csv("/kaggle/working/predicted_biomass_2016.csv", index=False)

print(final_2016)
```
