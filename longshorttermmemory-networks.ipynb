```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

```

```python

data = pd.read_csv("/kaggle/input/biomass-dist/Biomass_History.csv")


features = data[['Latitude', 'Longitude']]
target = data[['2010', '2011', '2012', '2013', '2014', '2015', '2016']]

X = features.values
y = target.values

scaler = StandardScaler()
X = scaler.fit_transform(X)


# Reshape y to a 3D array (samples, time steps, features)
y = y.reshape(-1, 7, 1)  


import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# modell
model = Sequential()
model.add(LSTM(64, input_shape=(7, 1)))  # 7 time steps, 1 feature (biomass generation)
model.add(Dense(1))  # Output layer

# Compile the modell
model.compile(loss='mse', optimizer='adam')

# Train the modell
model.fit(y, y[:, -1], epochs=100, batch_size=16)  # Predicting the last year (2017)

# Predict biomass generation for the year 2017 (last time step)
#predicted biomass_2017 = model.predict(y[:, -1:])
predicted_biomass_2017 = model.predict(y[-1].reshape(1, 7, 1))
# Predict Biomass Generation for the entire 2017
num_years = y.shape[1]  # Number of years in the historical data
predictions_2017 = []

for i in range(len(y)):
    input_sequence = y[i:i+1, :, :]  # Single input sequence
    prediction = model.predict(input_sequence)[0, 0]  # Get the predicted value
    predictions_2017.append(prediction)
predictions_2017 = np.array(predictions_2017)
print(predictions_2017)

```
